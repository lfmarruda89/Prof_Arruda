```python
import numpy as np # biblioteca para trabalhar com equações matemáticas ou valores não tradicionais
import matplotlib.pyplot as plt # biblioteca para plotar gráficos
```

```python
# Definindo o tempo
t = np.linspace(-2 * np.pi, 2 * np.pi, 1000) # intervalo do sinal

# sinal original
x = np.sin(t)

# Sinais escalados
alpha = 2
x_compressao = np.sin(alpha * t)  # Compressão

alpha2 = 3
x_expansao = np.sin(t / alpha2)  # Expansão

# Configuração da janela de impressão 'tamanho do grafico'
plt.figure(figsize=(12, 8))

# Plot do sinal original
plt.subplot(3, 1, 1) # pegar primeira janela de 3
plt.plot(t, x, label='Sinal Original: $x(t)$') # imprimir o sinal com legenda
plt.title('Sinal Original') # colocar um título na janela
plt.xlabel('Tempo $t$') # legenda no eixo x "tempo"
plt.ylabel('Amplitude') # legenda no eixo y "amplitude"
plt.legend() # habilitar impressão da legenda do sinal
plt.grid(True) # habilitar linhas de gride

# Plot do sinal comprimido
plt.subplot(3, 1, 2) # pegar segunda janela de 3
plt.plot(t, x_compressao, label='Sinal Comprimido: $x(2t)$', color='r') # imprimir o sinal com legenda
plt.title('Sinal Comprimido') # colocar um título na janela
plt.xlabel('Tempo $t$') # legenda no eixo x "tempo"
plt.ylabel('Amplitude') # legenda no eixo y "amplitude"
plt.legend() # habilitar impressão da legenda do sinal
plt.grid(True) # habilitar linhas de gride

# Ploat sinal expandido
plt.subplot(3, 1, 3) # pegar terceira janela de 3
plt.plot(t, x_expansao, label='Sinal Expandido: $x(0.5t)$', color='g') # imprimir o sinal com legenda
plt.title('Sinal Expandido') # colocar um título na janela
plt.xlabel('Tempo $t$') # legenda no eixo x "tempo"
plt.ylabel('Amplitude') # legenda no eixo y "amplitude"
plt.legend() # habilitar impressão da legenda do sinal
plt.grid(True) # habilitar linhas de gride

#plotar grafico
plt.tight_layout()
plt.show()
```
