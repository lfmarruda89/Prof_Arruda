```python
import numpy as np # biblioteca para trabalhar com equações matemáticas ou valores não tradicionais
import matplotlib.pyplot as plt # biblioteca para plotar gráficos
```

```python
# Parâmetros do sinal
frequencia = 60  # Frequência do sinal em Hz

T = 1/60  # Duração do sinal em segundos
```

```python
fs = 10000  # Frequência de amostragem em Hz

n_samples = int(T * fs)  # Número de amostras

# Definindo o tempo
t = np.linspace(0, T, n_samples, endpoint=False)  # Vetor de tempo discretizado

# sinal original
x = np.sin(2 * np.pi * frequencia * t)

# Plot do sinal original
plt.plot(t, x, label='Sinal Original: $x(t)$') # imprimir o sinal com legenda
plt.title('Sinal Original') # colocar um título na janela
plt.xlabel('Tempo $t$') # legenda no eixo x "tempo"
plt.ylabel('Amplitude') # legenda no eixo y "amplitude"
plt.legend() # habilitar impressão da legenda do sinal
plt.grid(True) # habilitar linhas de gride
#plotar grafico
plt.tight_layout()
plt.show()

```

```python
fs = 1000  # Frequência de amostragem em Hz

n_samples = int(T * fs)  # Número de amostras

# Definindo o tempo
t2 = np.linspace(0, T, n_samples)#, endpoint=False)  # Vetor de tempo discretizado
 
# Sinais amostrado
x_amostrado = np.sin(2 * np.pi * frequencia * t2)

plt.plot(t, x, label='Sinal Original: $x(t)$') # imprimir o sinal com legenda
plt.scatter(t2, x_amostrado, label='Sinal Amostrado: $x(t)$', color='r') # imprimir o sinal com legenda
plt.title('Sinal Discreto') # colocar um título na janela
plt.xlabel('Tempo $t$') # legenda no eixo x "tempo"
plt.ylabel('Amplitude') # legenda no eixo y "amplitude"
# plt.legend() # habilitar impressão da legenda do sinal
plt.grid(True) # habilitar linhas de gride

#plotar grafico
plt.tight_layout()
plt.show()
```
